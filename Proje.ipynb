```python
import re
import string

def text_temizle1(text):
    #kucuk harf yap
    text = text.lower()
    #parantez içindeki yazıları sil
    text = re.sub('\[.*?\]', '', text)
    #noktalama işaretlerini sil
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    #sayıları sil
    text = re.sub('\w*\d\w*', '', text)
    #3 karakterden kısa olan sözcükleri sil türkçe karakterlerle birlikte
    text = re.sub(r'\b[a-zğüşöçıİĞÜŞÖÇ]{1,2}\b', r'',text)
    return text

```

```python
#temizleme1 fonksyonuna gönder
data_clean = pd.DataFrame(dfTrimmed.Review.apply(text_temizle1))
```

```python
#Türkçe'de olumsuzluk, fiil + me/ma .. ile olur. Türkçede en çok kullanılan fiilerin listesini sırayla dolaşıp gelen 
# veriyle alakası olup olmadığını kontrol ederek olumsuzluğu tespit etmiş oluruz
def isnegative(text):
    count = 0
    for fiil in turkceFiiller:
        pattern = re.compile(r'%s(?=m(a|e|ı|i|o|ö|u|ü))' % fiil)
        results = pattern.finditer(text)
        
        for result in results:
            count = count + 1
        
        if count == 0:
            sonuc = False
        else:
            sonuc = True
            
    return sonuc
```

```python
sonuc = isnegative("gelmedi")
sonuc
```

```python
# her cümledeki her kelimeye bakıp negatif kelime bulduğunda cümlenin sonuna "kötü" ekliyoruz
for sentence in tokenized:
    for word in sentence:
        if isnegative(word) == True:
            sentence.append("kötü")
```
